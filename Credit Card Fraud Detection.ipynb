```python
#Build a sample dataset containing similar distributn of normal txns andd fraud txns

# Number of Fraud txns --> 492
# we'll take random 492 legit samples from 284315 and take them with 492 fraud samples
```

```python
legit_sample = legit.sample(n=492)  #randomly 492 datapoints not the first 492
```

```python
#Concatenating two dataframes
```

```python
new_dataset = pd.concat([legit_sample, fraud], axis= 0)  
#we want to have it row wise so mentioned axis 0 ie first 492 legit sample then 492 fraud
#if axis 1-> columnwise 
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
#the mean of classes of new_dataset is almost similar to dataset 
#So, nature of dataset has not chnged
# so diff is still there -> it creates our machine learning model able to detect whether our dtapoint is legit or fraud
```

```python
#SPLITTING THE DATA INTO FEATURES AND TARGETS
#features -> v1,v2,etc (X)
#targets -> 0 or 1     (Y)
```

```python
X = new_dataset.drop(columns='Class', axis= 1)
Y = new_dataset['Class']
```

```python
print(X)  #doesnt have class column as it was dropped
```

```python
print(Y)  # class column
```

```python
#SPLIT THE DATA INTO TRAINING DATA AND TESTING DATA
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y , test_size =0.2, stratify=Y, random_state=2)

#test.size=0.2 means 20% data -> testing ::: 80% data -> training ---- X_test stores 20% of data and Y_test stores its labels, X_train stores 80% of data and Y_train stores its labels
#stratify(Y) makes sure the Y ie classes ie 0 and 1 are evenly distributed in training and testing data
#random_state=2 --- means how you want to split your data, if we mention 2 and tutor also mentioned 2 , our datas will be splitted in the same way, if I mention 3 instead, my data will be splitted in a different way
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
#MODEL TRAINING
```

```python
#LOGISTIC REGRESSION (we're using this cuz it is generally used for binary classification
```

```python
model = LogisticRegression()
```

```python
#training the Logistic Regression Model with Training Data
model = LogisticRegression(max_iter=1000)

model.fit(X_train, Y_train)
```

```python
#Model Evaluation 
```

```python
#Accuracy Score
```

```python
#accuracy on training data
X_train_prediction = model.predict(X_train) #all the labels predicted for X_train which is stored in X_train_prediction
training_data_accuracy = accuracy_score(X_train_prediction, Y_train) #the labels predicted for X_train ie X_train_prediction are compared with the actual labels for X_train ie Y_train
#comparison gives us accuracy score which is stored in  training_data_accuracy
```

```python
print('Accuracy on Training Data:', training_data_accuracy)
```

```python
# if accuracy score > 73%or 80% --> good accuracy score
```

```python
#accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on Test Data:', test_data_accuracy)
```

```python
#so Accuracy on Test Data almost simlar to Accuracy on Training Data
```
